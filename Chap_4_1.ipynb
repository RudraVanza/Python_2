1> Adding New Column in Data Fram

```python
# 1> Adding New Column in Data Fram
import pandas as pd
length = [18,20,10,12,18,11]
width = [20,20,10,11,19,20]
data = {"length":length,"width":width}
df = pd.DataFrame(data)
df
df["area"] = df["length"]*df["width"] #For Adding New column
df
```

2> Convert Categorical Data into numeric data 

```python
age = [18,20,23,19,18,22]
city = ["A","B","B","A","C","B"]
data = {"age":age,"city":city}
df = pd.DataFrame(data)
df
df = pd.get_dummies(data=df) #To Get Dummi Data , Converting Numeric Data,, drop_first=True => For 
df 
```

```python
df = pd.read_csv("car data.csv")
df = df.drop("Car_Name",axis=1) # To Remove Car_Name Column
df = pd.get_dummies(data=df) # To Get Dummies Of All Column
df
```
