# Simple Linear Regression
Steps :- 
    1> Load Data Set    
    2> Check Null Value         
    3> Shape Of Data Frame          
    4> Define x & y             
    5> Print Shape OF x & y                 
    6> Split Data Into Train & Test                 
    7> Model Selection                  
    8> Find m & c                   
    9> Find Error                

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Step 1 :- Load Data Set
df = pd.read_csv("Book1.csv")
df

#Step 2 :- Check Null Value
df.isna().sum()

# Step 3 :- Shape Of Data Frame
df.shape

# Step 4 :- Define x & y
x = df[['cgpa']]
x
y = df[['package']]
y

# Step 5 :-  Print Shape OF x & y
x.shape
y.shape

# Step 6 :- Split Data Into Train & Test
# from sklearn.model_selection import train_test_split => For Step 6
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
x_train.shape #(160,1)
x_test.shape #(40,1)
y_train.shape #(160,1)
y_test.shape #(40,1)
''
''
# Step 7 :- Model Selection
# from sklearn.linear_model import LinearRegression => for Step 7
lr = LinearRegression()
lr.fit(x_train,y_train) # To Train Model

# Step 8 :- Find m & c   
m = lr.coef_ # Find m for y = mx + c
print("Value Of m :- ",m)
c = lr.intercept_ # Find c for y = mx + c , intercept_ = Slope
print("Value Of c :- ",c)

# y1 = lr.predict([[7]]) # For Show For & CGPA
# print("Packege of 7 CGPA :- ",y1)

# Step 9 :- Find Error   
# For Find ERROR in Model :- from sklearn.metrics import mean_squared_error
print(y_test)
y_pred = lr.predict(x_test) # To Find Predicted Packege Of x_test
print(y_pred)
print("mean_squared_error(MSE) :- ",mean_squared_error(y_test,y_pred)) # mean_squared_error(y_test,y_pred) => To find mean_squre_error
```

# Multiple Linear Regression
Formula :- " y = m1x1 + m2x2 + m3x3 + .... + c "                          
Steps:-                 
    1> Load Data             
    2> Print Shape              
    3> Check null Value                 
    4> Data Preprocessing                   
    5> Data Preparation                         
    6> Define x & y                                         
    7> Split data into Train & Test                                                         
    8> Model Selection                                    
    9> Find m & c                    
    10> Find Error                

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score , mean_absolute_error

# Step 1> Load Data  
df = pd.read_csv("car data.csv")
df

# Step 2> Print Shape   
print(df.shape)

# Step 3> Check null Value
print(df.isna().sum())

# Step 4> Data Preprocessing 
df["Age"] = 2025 - df["Year"]  # To get age of car
df
df = df.drop("Year",axis=1) # to drop year column
df = df.drop("Car_Name",axis=1) # to drop Car_Name column

# Step 5> Data Preparation 
df = pd.get_dummies(data=df,drop_first=True) #To get dummy columns and drop first dummy column
df

df.columns  # To get all column name from data frame

# Step 6> Define x & y
y = df["Selling_Price"]
print(y)
x = df.drop("Selling_Price",axis=1)
print(x)
print("Shape of x :- ",x.shape)
print("Shape of y :- ",y.shape)

# Step 7> Split data into Train & Test
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2,random_state=2)
print("X_Train :- ",x_train.shape)
print("X_Test :- ",x_test.shape)
print("Y_Train :- ",y_train.shape)
print("Y_Test :- ",y_test.shape)

# Step 8> Model Selection
lr = LinearRegression()
lr.fit(x_train,y_train)

# Step 9> Find m & c
m = lr.coef_
print("value of m :- ",m)
c = lr.intercept_
print("Value of c :- ",c) 

# Present price 5.5, Kms_Driven = 100000, Owner = 0, Year = 10, Fuel_Type = Petrol , Seller_Type = indivisual , Transmission = menual
y1 = lr.predict([[5.5,100000,0,10,0,1,0,1]])
print("Predicted Price is (y1) : ",y1)

# Present price = 7 , Kms_Driven = 90000, Owner = 0, Year = 5, Fuel_Type = CNG , Seller_Type = indivisual , Transmission = Auto
y2 = lr.predict([[7,90000,0,5,0,0,1,0]])
print("Predicted Price is (y2) : ",y2)

# Step 10> Find Error
y_pred = lr.predict(x_test) # To Find Predicted Price Of Car 
print("MSE :- ",mean_squared_error(y_test,y_pred))
print("RMSE :- ",mean_squared_error(y_test,y_pred)**(0.5))
print("MAE :- ",mean_absolute_error(y_test,y_pred))
print("R2_score :- ",r2_score(y_test,y_pred))
```
