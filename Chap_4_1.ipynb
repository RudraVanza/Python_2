# Simple Linear Regression
Steps :- 
    1> Load Data Set    
    2> Check Null Value         
    3> Shape Of Data Frame          
    4> Define x & y             
    5> Print Shape OF x & y                 
    6> Split Data Into Train & Test                 
    7> Model Selection                  
    8> Find m & c                   
    9> Find Error                

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Step 1 :- Load Data Set
df = pd.read_csv("Book1.csv")
df

#Step 2 :- Check Null Value
df.isna().sum()

# Step 3 :- Shape Of Data Frame
df.shape

# Step 4 :- Define x & y
x = df[['cgpa']]
x
y = df[['package']]
y

# Step 5 :-  Print Shape OF x & y
x.shape
y.shape

# Step 6 :- Split Data Into Train & Test
# from sklearn.model_selection import train_test_split => For Step 6
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
x_train.shape #(160,1)
x_test.shape #(40,1)
y_train.shape #(160,1)
y_test.shape #(40,1)
''
''
# Step 7 :- Model Selection
# from sklearn.linear_model import LinearRegression => for Step 7
lr = LinearRegression()
lr.fit(x_train,y_train) # To Train Model

# Step 8 :- Find m & c   
m = lr.coef_ # Find m for y = mx + c
print("Value Of m :- ",m)
c = lr.intercept_ # Find c for y = mx + c , intercept_ = Slope
print("Value Of c :- ",c)

# y1 = lr.predict([[7]]) # For Show For & CGPA
# print("Packege of 7 CGPA :- ",y1)

# Step 9 :- Find Error   
# For Find ERROR in Model :- from sklearn.metrics import mean_squared_error
print(y_test)
y_pred = lr.predict(x_test) # To Find Predicted Packege Of x_test
print(y_pred)
print("mean_squared_error(MSE) :- ",mean_squared_error(y_test,y_pred)) # mean_squared_error(y_test,y_pred) => To find mean_squre_error
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


df = pd.read_csv("olympic100.csv")
df

print(df.isna().sum())

print(df.shape)

x = df[['year']]
print(x)
y = df[['time']]
print(y)

print("Shape of x :- ",x.shape)
print("Shape of y :- ",y.shape)

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2,random_state=2)
print("X_Test :- ",x_test)
print("X_Train :- ",x_train)
print("T_Test :- ",y_test)
print("X_Train :- ",y_train)

lr = LinearRegression()
lr.fit(x_train,y_train)

m = lr.coef_
print(m)
c = lr.intercept_
print(c)

y1 = lr.predict([[2025]])
print("Y1 :- ",y1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


df = pd.read_csv("heights_weights.csv")
df

print(df.isna().sum())

print(df.shape)

x = df[['Height']]
print(x)
y = df[['Weight']]
print(y)

print("Shape of x :- ",x.shape)
print("Shape of y :- ",y.shape)

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2,random_state=2)
print("X_Test :- ",x_test)
print("X_Train :- ",x_train)
print("T_Test :- ",y_test)
print("X_Train :- ",y_train)

lr = LinearRegression()
lr.fit(x_train,y_train)

m = lr.coef_
print(m)
c = lr.intercept_
print(c)

y1 = lr.predict([[80]])
print("Y1 :- ",y1)

print(y_test)
y_pred = lr.predict(x_test) # To Find Predicted Packege Of x_test
print(y_pred)
print("mean_squared_error(MSE) :- ",mean_squared_error(y_test,y_pred))
```
