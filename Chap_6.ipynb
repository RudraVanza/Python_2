```python
#for categorical data
#1.knn algorithm..k-nearest neighbour
import pandas as pd
df=pd.read_csv('diabetes.csv')
df
y=df[['Outcome']]
y
x=df.drop('Outcome',axis=1)
x
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
from sklearn.neighbors import KNeighborsClassifier
nn=KNeighborsClassifier(n_neighbors=27)
model=nn.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
y_test
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

1> Accuracy :- TP + TN / TN + TP + FN + FP                                                  
2> Error Rate :- FP + FN / TN + TP + FN + FP    OR   1 - Accuracy                           
3> Sensitivity :- TP / TP + FN                                                              
4> Specificity :- TN / TN + FP                                                               

```python
df = pd.read_csv("PlayTennis.csv")
print(df,"\n")
df = pd.get_dummies(df,drop_first=True)
print(df,"\n")
y = df[["play_yes"]]
print("Y :- ",y,"\n")
x = df.drop("play_yes",axis=1)
print("X :- ",x,"\n")
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
print("x_train :- ",x_train.shape)
print("y_train :- ",y_train.shape)
print("x_test :- ",x_test.shape)
print("y_test :- ",y_test.shape,"\n")
nn=KNeighborsClassifier(n_neighbors=3)
model=nn.fit(x_train,y_train)
y_pred=model.predict(x_test)
print("y_pred :- ",y_pred,"\n")

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print("cm :- ",cm,"\n")

tn = cm[0][0]
fp = cm[0][1]
tp = cm[1][1]
fn = cm[1][0]
print(f"TN :- {tn} \nTP :- {tp} \nFN :- {fn} \nFP :- {fp}")

Accuracy = (tn+tp)/(tn+tp+fn+fp)
Error_Rate = 1 - Accuracy
Sensitivity = tp / (tp + fn)
Specificity = tn / (tn + fp) 
print(f"Accuracy :- {Accuracy}\nError_Rate :- {Error_Rate}\nSensitivity :- {Sensitivity}\nSpecificity :- {Specificity}")
```

```python
# 1> Total Charges column fill with mean Value
# 2> drop :- customerID , Unnamed: 0	
# 3> define x = other all  , y = Churn
# 4> nebour = 5, rendom state = 2
df = pd.read_csv("Churn.csv")
df["TotalCharges"] = df["TotalCharges"].fillna(df["TotalCharges"].mean())
df
df = df.drop(["Unnamed: 0","customerID"],axis=1)      
df = pd.get_dummies(df,drop_first=True)
df
y= df[["Churn_Yes"]]
x = df.drop("Churn_Yes",axis=1)
print(f"X :- {x}\nY :- {y}")

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
print("x_train :- ",x_train.shape)
print("y_train :- ",y_train.shape)
print("x_test :- ",x_test.shape)
print("y_test :- ",y_test.shape,"\n")

from sklearn.neighbors import KNeighborsClassifier
nn=KNeighborsClassifier(n_neighbors=5)
model = nn.fit(x_train,y_train)
model
y_pred=model.predict(x_test)
print("y_pred :- ",y_pred,"\n")

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(f"cm :- {cm}\n")

tn = cm[0][0]
fp = cm[0][1]
tp = cm[1][1]
fn = cm[1][0]
print(f"TN :- {tn} \nTP :- {tp} \nFN :- {fn} \nFP :- {fp}")

Accuracy = (tn+tp)/(tn+tp+fn+fp)
Error_Rate = 1 - Accuracy
Sensitivity = tp / (tp + fn)
Specificity = tn / (tn + fp) 
print(f"Accuracy :- {Accuracy}\nError_Rate :- {Error_Rate}\nSensitivity :- {Sensitivity}\nSpecificity :- {Specificity}")
```

```python
# 1> load data set Fish.csv 
# 2> Dependent var waight = y , x = length 1,2,3,hight  
# 3> rendom state = 25 , 
# 4 > polynomial regeressinon degree 3
# 5> r2_score , mean_squre_error 
# 6> test case :- l1 = 20 , l2 = 20 , l3 = 30 , hight = 12 , widrh = 5 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score , mean_absolute_error
from sklearn.preprocessing import PolynomialFeatures 
df = pd.read_csv('Fish.csv')
df.isna().sum()
y = df[["Weight"]]
x = df.drop(["Species","Weight"],axis=1)
print(f"X :- {x}\nY :- {y}")
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2,random_state=42)
print("X_Train :- ",x_train.shape)
print("X_Test :- ",x_test.shape)
print("Y_Train :- ",y_train.shape)
print("Y_Test :- ",y_test.shape) 
print(df)

poly = PolynomialFeatures(degree=3)
x_poly = poly.fit_transform(x_train)
print(f"x_poly :- {x_poly}")

lr = LinearRegression()
lr.fit(x_poly,y_train)

m = lr.coef_
print("m :- ",m)
c = lr.intercept_ 
print("c :- ",c)

x_poly_test = poly.fit_transform(x_test)

y_pred = lr.predict(x_poly_test)
print(f"y_pred :- {y_pred}")
print(f"R2_SCORE :- {r2_score(y_test,y_pred)}\nmean_squre_error :- {mean_squared_error(y_test,y_pred)}")
```
