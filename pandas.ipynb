```python
from wordcloud import WordCloud,STOPWORDS
f=open("11-0.txt",encoding="utf-8")
alice_novel=f.read()
print(alice_novel)
```

```python
stop=set(STOPWORDS)
stop
```

```python
alice_wd=WordCloud(background_color="white",max_words=2000,stopwords=stop)
print(alice_wd)
```

```python
alice_wd.generate(alice_novel)
```

```python
#display
import matplotlib.pyplot as plt
plt.figure(figsize=(14,18))
plt.imshow(alice_wd)
plt.axis("off")
plt.show()
```
